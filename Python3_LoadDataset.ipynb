```python
def feet_inch(h):
  feet,inches = list(map(int,h.split("-")))
  return (((feet*12 + inches) * 2.54)/100)
```

```python
nba=pd.read_csv('nba.csv')
nba.loc[111,"Height"]
```

```python
#df_titanic["Revised_Fare"]= df_titanic.apply(lambda x : x["Fare"]*100,axis=1)
#df_titanic["ClasswiseFare"]= df_titanic.apply(lambda x : x["Fare"]+120/x["Class"],axis=1)
#def Farecalulator(fare,cls) :
 # return fare + (120/cls)
#df_titanic["ClasswiseFare"]= df_titanic.apply(lambda x : Farecalulator(x["Fare"],x["Class"]),axis=1)
nba["Height_m"]=nba.apply(lambda x : feet_inch(x["Height"]),axis=1)
nba
nba["Weight_Kg"]=nba.apply(lambda x : x["Weight"]*0.4535,axis=1)
nba
nba["BMI"]=nba.apply(lambda x :round( x["Weight_Kg"]/x["Height_m"]**2,2),axis=1)
nba
```

```python
def BMI_check (a):
  if a<= 18.4:
    return "Underweight"
  elif a >= 18.4 and a <=24.9:
    return "Healthy"
  elif a>24.0 and a <=29.9:
    return "Over Weight"
  elif a >29.9 and a <= 34.9:
    return "Severly Over Weight"
  elif a >34.9 and a <= 39.9:
    return "Obese"
  else:
    return "Severly Obese"
```

```python
nba["BMI_cat"]=nba.apply(lambda x: BMI_check(x["BMI"]),axis=1)
nba["BMI_cat"].value_counts()
```

```python
#Compute the total amount paid by each team to the players.
#What is the amount paid by Chicago Bulls to the players?
nba.groupby("Team").sum()
```

```python
#Compute the average age of each team by their playing position. What is the average age of the Utah Jazz team players playing at the ‘PG’ position?
nba.groupby(["Team", "Position"])["Age"].mean().loc["Utah Jazz"]
```

```python
#Compute each team's average, maximum, and minimum age by their playing position.
#What is the average, maximum, and minimum age of the Atlanta Hawks players playing at the ‘PG’ position?
nba.groupby(["Team", "Position"])["Age"].agg(["max","mean","min"]).loc["Atlanta Hawks"]

```

```python
#Compute each team's average age and total salary by their playing position.
#Sort the grouped results based on salary.
#Enter the team name with the highest salary.
nba.groupby(by=["Team", "Position"]).agg({"Age" : "mean","Salary" : "sum"}).sort_values(by="Salary", ascending=False)
```
