Series to DataFrame

```python
name=["Adsa","das","qwdd","edas"]
values=[100,210,2453,21432]
nameser=pd.Series(name)
valueser=pd.Series(values)
df=pd.DataFrame([nameser,valueser])
df_new=df.transpose()
df_new.columns=["Name","Values"]
df_new
df_another_way=pd.DataFrame({"Name":nameser,"Values":valueser})
df_another_way
```

Setting Index in DataFrame


```python
name=["Adsa","das","qwdd","edas"]
values=[100,210,2453,21432]
#indexes=[a,b,c,d]
nameser=pd.Series(name)
valueser=pd.Series(values)

df=pd.DataFrame({"Name": nameser,"Values": valueser,"indexes": ["a","b","c","d"]})
df
```
