```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

def data_preprocessing_pipeline(data):
    # Create a copy of the original DataFrame to avoid modifying the input data
    data_copy = data.copy()

    # Identify numeric and categorical features
    numeric_features = data_copy.select_dtypes(include=['float', 'int']).columns
    categorical_features = data_copy.select_dtypes(include=['object']).columns

    # Handle missing values in numeric features
    data_copy[numeric_features] = data_copy[numeric_features].fillna(data_copy[numeric_features].mean())

    # Detect and handle outliers in numeric features using IQR
    for feature in numeric_features:
        Q1 = data_copy[feature].quantile(0.25)
        Q3 = data_copy[feature].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - (1.5 * IQR)
        upper_bound = Q3 + (1.5 * IQR)
        data_copy[feature] = np.where((data_copy[feature] < lower_bound) | (data_copy[feature] > upper_bound),
                                      data_copy[feature].median(), data_copy[feature])


    # Handle missing values in categorical features
    data_copy[categorical_features] = data_copy[categorical_features].fillna(data_copy[categorical_features].mode().iloc[0])

    return data_copy

# Create a sample dataset with outliers
import pandas as pd
import numpy as np

# Set a random seed for reproducibility
np.random.seed(42)

# Create a sample DataFrame with weights of boys and girls
data = pd.DataFrame({
    'Gender': ['Boy'] * 50 + ['Girl'] * 50,
    'Weight': np.concatenate([
        np.random.normal(loc=50, scale=5, size=47),     # Boys' weights (normal distribution)
        np.random.normal(loc=45, scale=5, size=47),     # Girls' weights (normal distribution)
        [200, 180, 210],                               # Introduce some outlier values for boys
        [35, 30, np.nan],                              # Introduce some missing values for girls
    ])
})

# Shuffle the DataFrame to mix boys and girls data
data = data.sample(frac=1, random_state=42).reset_index(drop=True)

# Display the DataFrame with outliers and missing values


data
```
